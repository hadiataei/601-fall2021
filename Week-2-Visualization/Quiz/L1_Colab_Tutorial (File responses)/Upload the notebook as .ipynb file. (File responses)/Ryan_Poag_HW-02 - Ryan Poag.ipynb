Please complete the functions below. These functions are to simulate a fair and a biased die. 

The biased die has probabilities {0.25, 0.15, 0.15, 0.15, 0.15, 0.15}.

Create 2 visualizations that shows outcomes of multiple rolls of a fair die and biased die. 
* Visualization for fair die
* Visualization for biased die

Repeat the process programatically and take the mean to create a new list. This list will be used to compare means of biased & unbiased dice. 
* Create a historgram showing means of biased & unbiased die

* Use subcharts
* All visualizations must have appropriate titles.
* There are 3 visualization and expected format is 
```
 [V1] [V2]
 [   V3  ]
 ```

````python
import random 
import matplotlib.pyplot as plt

trial_count = 500

def fair_die_simulation(trials):
    """
    returns an array of randomly numbers between 1 and 6. 
    """
    
    options = [1,2,3,4,5,6]
    simulations = random.choices(options, k=trials)
    return simulations
````

````python
def biased_die_simuations(trials, weights=[0.25, 0.15, 0.15, 0.15, 0.15, 0.15]):
    """
    returns an array of randomly numbers between 1 and 6 with the probabilty of assosiated weight. 
    for instance: for weights: [0.25, 0.15, 0.15, 0.15, 0.15, 0.15]
    1 has 0.25 change to be selected, 2 has 0.15 chance and so on...
    """
    
    options = [1,2,3,4,5,6]
    simulations = random.choices(options, weights=weights, k=trials)
    return simulations
````

````python
biased = biased_die_simuations(trial_count)
fair = fair_die_simulation(trial_count)
````

````python
fig, (ax1, ax2) = plt.subplots(1, 2, sharex=True, sharey=True, figsize=(12,6))

ax1.hist(biased)
ax1.set_title('Biased')
ax1.set_xlabel('Die Value')
ax1.set_ylabel('# Occurances')

ax2.hist(fair)
ax2.set_xlabel('Die Value')
ax2.set_title('Fair')

fig.suptitle('Die roll comparison')
fig.show()
````

````python
i=0
biased_mean_list = []
unbiased_mean_list = []

while i < trial_count:
  biased = biased_die_simuations(trial_count)
  fair = fair_die_simulation(trial_count)
  biased_mean_list.append(sum(biased)/len(biased))
  unbiased_mean_list.append(sum(fair)/len(fair))
  i += 1
````

````python
grid = plt.GridSpec(2,2, wspace=.4, hspace=.6)
fig = plt.figure(figsize=[8,6])

ax1 = plt.subplot(grid[0,0])
ax1.hist(biased_mean_list)
ax1.set_title('Biased')
ax1.set_xlabel('Mean')
ax1.set_ylabel('# Occurances')

ax2 = plt.subplot(grid[0,1])
ax2.hist(unbiased_mean_list)
ax2.set_xlabel('Mean')
ax2.set_title('Fair')

ax3 = plt.subplot(grid[1,:])
ax3.hist([biased_mean_list, unbiased_mean_list])
ax3.set_title('Biased vs Unbiased')
ax3.legend(['Biased', 'Unbiased'])
ax3.set_xlabel('Mean')
ax3.set_ylabel('# Occurances')

fig.suptitle('Die roll trial mean comparison')
fig.show()
````

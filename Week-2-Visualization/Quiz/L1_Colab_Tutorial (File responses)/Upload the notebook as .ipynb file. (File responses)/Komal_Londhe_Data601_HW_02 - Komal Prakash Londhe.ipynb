````python
import random 
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
sns.set()

trial_count = 500
fair_die_simulation_array = []
unbiased_mean_list = []
def fair_die_simulation(trials):  #returns an array of randomly numbers between 1 and 6.
    options = [1,2,3,4,5,6]
    simulations = []
    for i in range(trials):
      x = random.randint(options[0],options[5])
      simulations.append(x)
    #print(simulations.count(options[0]))
    for i in range(len(options)):
      unbiased_mean_list.append(simulations.count(options[i])/trials)
    print("No of trial count is : ", len(simulations))
    plt.hist(simulations, bins=len(options), align='mid')
    plt.xlabel("Number on a fair die")
    plt.ylabel("Number of simulations")
    plt.title("Fair Die Simulation")
    plt.show()
    return simulations

fair_die_simulation_array = fair_die_simulation(trial_count)
print("\n Fair die Simulations are : \n", fair_die_simulation_array)
print("\n Mean values of a fair die : \n", unbiased_mean_list)
````

````python
from collections import Counter
import numpy as np
import random
from random import random 

trial_count = 500
die_weights = [0.25, 0.15, 0.15, 0.15, 0.15, 0.15]
biased_die = []
biased_mean_list = []
def weighted_choice(options, weights):
    weights = np.array(weights, dtype=np.float64)
    sum_of_weights = weights.sum()
    # standardization:
    np.multiply(weights, 1 / sum_of_weights, weights)
    weights = weights.cumsum()
    x = random()
    for i in range(len(weights)):
        if x < weights[i]:
            return options[i]

def biased_die_simuations(trials, weights=[0.25, 0.15, 0.15, 0.15, 0.15, 0.15]):
    options = [1,2,3,4,5,6]
    simulations = []
    for i in range(trials):
      simulations.append(weighted_choice(options, weights))
    for i in range(len(options)):
      biased_mean_list.append(simulations.count(options[i])/trials)
    return simulations

biased_die = biased_die_simuations(trial_count,die_weights)
print("\n Number of trial count is : ", len(biased_die))
print("\n Biased die Simulations are : \n", biased_die)
cnt = Counter(biased_die)
for key,values in cnt.items():
    cnt[key] = cnt[key] / trial_count  
    print("\nFace value on the die : "+str(key),"    Number of times "+str(key)+" occured : "+str(values))
print("\nThe options are : ", cnt.keys())
print("\nProbilities with respect to options are : ", cnt.values())
print("\nSum of probability is : ", sum(cnt.values()))
print("\n Mean values of a biased die : \n", biased_mean_list)
print("\n\n")

plt.bar(cnt.keys(),cnt.values(),width=0.75)
plt.xlabel("Number on a biased die")
plt.ylabel("Probability")
plt.title("Biased Die Simulation")
plt.show()
````

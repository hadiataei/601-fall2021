````python
import random 
import matplotlib.pyplot as plt
````

````python

trial_count = 500

def fair_die_simulation(trials):
    """
    returns an array of randomly numbers between 1 and 6. 
    """
    simulations = []
    for x in range (1,trials):
      options = [1,2,3,4,5,6]
      number = random.randint(1,6)
      simulations.append(number)
    return simulations
````

````python

fair = fair_die_simulation(trial_count)
````

````python

options = [1,2,3,4,5,6]
unbiased_mean_list = []
unbiased_mean_list.append(fair.count(1))
unbiased_mean_list.append(fair.count(2))
unbiased_mean_list.append(fair.count(3))
unbiased_mean_list.append(fair.count(4))
unbiased_mean_list.append(fair.count(5))
unbiased_mean_list.append(fair.count(6))
````

````python
print(unbiased_mean_list)
````

````python
plt.xlabel('Die Roll')
plt.ylabel('Count')
plt.title('Bar Plot of Fair Dice')
plt.bar(options,unbiased_mean_list,align='center')
````

````python
plt.xlabel('Die Roll')
plt.ylabel('Mean Count')
plt.title('Histogram of Fair Dice')
plt.hist(fair,bins=6)
````

````python
def biased_die_simuations(trials, weights=[0.25, 0.15, 0.15, 0.15, 0.15, 0.15]):
    """
    returns an array of randomly numbers between 1 and 6 with the probabilty of assosiated weight. 
    for instance: for weights: [0.25, 0.15, 0.15, 0.15, 0.15, 0.15]
    1 has 0.25 change to be selected, 2 has 0.15 chance and so on...
    """
    
    options = [1,2,3,4,5,6]
    simulations = []
    simulations = random.choices(options, weights=weights, k=trials) # random.choices from https://www.geeksforgeeks.org/how-to-get-weighted-random-choice-in-python/
    return simulations
````

````python
biased = biased_die_simuations(trial_count,weights=[0.25, 0.15, 0.15, 0.15, 0.15, 0.15])
````

````python
options = [1,2,3,4,5,6]
biased_mean_list = []
biased_mean_list.append(biased.count(1))
biased_mean_list.append(biased.count(2))
biased_mean_list.append(biased.count(3))
biased_mean_list.append(biased.count(4))
biased_mean_list.append(biased.count(5))
biased_mean_list.append(biased.count(6))
````

````python
print(biased_mean_list)
````

````python
plt.xlabel('Die Roll')
plt.ylabel('Count')
plt.title('Bar Plot of Biased Dice')
plt.bar(options,biased_mean_list,align='center')
````

````python
plt.xlabel('Die Roll')
plt.ylabel('Mean Count')
plt.title('Histogram of Biased Dice')
plt.hist(biased,bins=6)
````

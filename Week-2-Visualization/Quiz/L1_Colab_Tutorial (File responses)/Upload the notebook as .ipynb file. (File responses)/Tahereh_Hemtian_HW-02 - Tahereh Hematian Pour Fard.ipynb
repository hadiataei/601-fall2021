````python
import random 
import matplotlib.pyplot as plt
trial_count = 500

def fair_die_simulation(trials):
  simulations = []
  for i in range(trials):
    rn = random.uniform(0,1)
    if 0 <= rn < 1/6:
      x = 1
    elif 1/6 <= rn < 2/6:
      x = 2
    elif 2/6 <= rn < 3/6:
      x = 3
    elif 3/6 <= rn < 4/6:
      x = 4
    elif 4/6 <= rn < 5/6:
      x = 5
    elif 5/6 <= rn <= 1:
      x = 6
    simulations.append(x)
  return simulations
simulations = fair_die_simulation(trial_count)
print(simulations)
options = [1,2,3,4,5,6]
plt.figure()
plt.title('Fair die simulation')
plt.xlabel('Die face')
plt.ylabel('Frequency')
plt.hist(simulations, bins= 6)
plt.show()
````

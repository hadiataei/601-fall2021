# Question-1

Write a Python function to returns a **list** of all the numbers from 1 to n which are divisible by 5. `n` will be given as an argument to your function.


```python
## Your Answer Cell
## you can edit the function below. 
## Don't change function name and parameter names.
## you can define new variables inside of the function.
def divisible_five(given_number = 1000):
  my_list = []
  for i in range(given_number):
    if i%5 == 0:
      my_list.append(i)
  return my_list
```

````python
import random 
import matplotlib.pyplot as plt

trial_count = 500

def fair_die_simulation(trials):
    """
    returns an array of randomly numbers between 1 and 6. 
    
    """
    simulations = []
    for i in range(trials):
      p = random.randrange(1,7)
      simulations.append(p)
    
    #options = [1,2,3,4,5,6]
    
    return simulations
````

````python
# part 3: simulate the the two and get the mean
rep = 50
mean_fair = [0]*rep
mean_biased = [0]* rep
for i in range(rep):
  

  fair_simulations = fair_die_simulation(trial_count)
  biased_simulations = biased_die_simuations(trial_count)
  mean_fair[i] = sum(fair_simulations)/len(fair_simulations)

  mean_biased[i] = sum(biased_simulations)/len(biased_simulations)
````
